```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
npts = 1000
#xx = np.random.rand(npts)*30-10
x1 = np.random.rand(npts)*10-15
x2 = np.random.rand(npts)*10-5
x3 = np.random.rand(npts)*10+5
rr = np.random.rand(npts)*2-1 # random scatter if required
yy = -30 + 2*x1 + 4*x2 + 6*x3 # m2*x2 + m3*x3

plt.scatter(x1,yy,marker='.',color='red')
plt.scatter(x2,yy,marker='.',color='green')
plt.scatter(x3,yy,marker='.',color='blue')
plt.show()
```

```python
x1mean = sum(x1)/npts
x2mean = sum(x2)/npts
x3mean = sum(x3)/npts
yymean = sum(yy)/npts

x1diff=0; x2diff=0; x3diff=0
y1diff=0; y2diff=0; y3diff=0

for i in range(0, npts-1):
     y1diff += (x1[i]-x1mean)*(yy[i]-yymean) #top
     y2diff += (x2[i]-x2mean)*(yy[i]-yymean) #top
     y3diff += (x3[i]-x3mean)*(yy[i]-yymean) #top
     x1diff += (x1[i]-x1mean)*(x1[i]-x1mean) #bot
     x2diff += (x2[i]-x2mean)*(x2[i]-x2mean) #bot
     x3diff += (x3[i]-x3mean)*(x3[i]-x3mean) #bot

m1 = (y1diff)/(x1diff)
m2 = (y2diff)/(x2diff)
m3 = (y3diff)/(x3diff)
c1 = yymean-(m1*x1mean)
c2 = yymean-(m2*x2mean)
c3 = yymean-(m3*x3mean)
cc = yymean - m1*x1mean - m2*x2mean - m3*x3mean

print 'm1 =',m1
print 'm2 =',m2
print 'm3 =',m3
print 'c1 =',c1
print 'c2 =',c2
print 'c3 =',c3
print 'cc =',cc
```
